```python
import tensorflow as tf
import numpy as np
import glob
import os
from scipy import misc
```

```python
train_dir = "./data/train/train_/"
save_dir = "./data/train/downsampled/"
```

```python
filenames = glob.glob(train_dir + "*.png")
```

```python
for filename in filenames:
    image = misc.imread(filename)
    resizedImage = misc.imresize(image,0.5)
    
    _,filename = os.path.split(filename)
    # save the resized image
    misc.imsave(save_dir + filename,resizedImage)
    print "file saved"
```
